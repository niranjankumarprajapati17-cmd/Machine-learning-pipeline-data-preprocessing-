# Retrieving Feature Names

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

df=pd.read_csv('spotify_global_trends_2026.csv')

# Define a numerical pipeline
df_pipeline = make_pipeline(SimpleImputer(strategy='median'), StandardScaler())

# Select the numerical columns from the spotify DataFrame for processing
# Based on common numerical columns in such datasets:
numerical_cols_spotify = ['streams', 'stream_change', '7day', 'days', 'viral_score']

# Apply the pipeline to transform the numerical data
data_transformed = df_pipeline.fit_transform(df[numerical_cols_spotify])

# Create a DataFrame from the transformed data with proper column names and index
df_streams_prepared=pd.DataFrame(
    data_transformed,
    columns=df_pipeline.get_feature_names_out(numerical_cols_spotify),
    index=df.index
)
```
